# DAPPS

### ¿Qué tecnologías son necesarias?
1) **Solidity:** Solidity es un lenguaje de programación orientado a Smart Contract
2) **Truffle:** Truffle, es un marco de desarrollo para Ethereum. Este nos ayudará a realizar operaciones como:
    - Compilación de smartContracts: **truffle compile**
    - Despliegue de contratos: **truffle migrate --network development**
    - Comprobación de funcionamiento mediante su consola: **truffle console**
3) **Ganache:** Ganache, es una herramienta proporcionada por la empresa Truffle, la cual nos permite tener una red local de Ethereum en nuestro ordenador

### ¿Qué son los Smart Contracts?
Un **Smart Contract**, es un programa que se ejecuta automáticamente en una BlockChain. 
Sus principales funciones son las siguientes: 
1) **Automatización:** No hay necesidad de intervención humana gracias a su programación basada en **eventos**. De esta forma se consigue eliminar la interpretación de las clausulas del contrato (funciones) ya que estas clausulas se realicen quedará registro de la misma.
2) **Transparencia:** Todas las transacciones quedan registradas en la red de la BlockChain.
3) **Seguridad:** Los Smart Contracts se ejecutan en un entorno virtual en la BlockChain. Y una vez desplegados ya no se pueden eliminar debido a la condición de Descentralización de este tipo de redes.
