# DAPPS

### Qu tecnologas son necesarias?
1) **Solidity:** Solidity es un lenguaje de programacin orientado a Smart Contract
2) **Truffle:** Truffle, es un marco de desarrollo para Ethereum. Este nos ayudar a realizar operaciones como:
    - Compilacin de smartContracts: **truffle compile**
    - Despliegue de contratos: **truffle migrate --network development**
    - Comprobacin de funcionamiento mediante su consola: **truffle console**
3) **Ganache:** Ganache, es una herramienta proporcionada por la empresa Truffle, la cual nos permite tener una red local de Ethereum en nuestro ordenador

### Qu son los Smart Contracts?
Un **Smart Contract**, es un programa que se ejecuta automticamente en una BlockChain. 
Sus principales funciones son las siguientes: 
1) **Automatizacin:** No hay necesidad de intervencin humana gracias a su programacin basada en **eventos**. De esta forma se consigue eliminar la interpretacin de las clausulas del contrato (funciones) ya que estas clausulas se realicen quedar registro de la misma.
2) **Transparencia:** Todas las transacciones quedan registradas en la red de la BlockChain.
3) **Seguridad:** Los Smart Contracts se ejecutan en un entorno virtual en la BlockChain. Y una vez desplegados ya no se pueden eliminar debido a la condicin de Descentralizacin de este tipo de redes.
